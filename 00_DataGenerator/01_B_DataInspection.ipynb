# Data Inspection

```python
import os
import arrow
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cwd = os.getcwd()
print('CWD: \t\t{}'.format(cwd))
data_fn = os.path.join(cwd, 'data_modelling')
fig_fn = os.path.join(cwd, 'figs')
print('Data: \t\t{}'.format(data_fn))
print('Figures: \t{}'.format(fig_fn))
```

```python
df_data = pd.read_csv(os.path.join(data_fn, 'main.csv'),sep=';', index_col='index')
```

```python
df_data.head(5)
```

```python
data = 'load_h0_normed'
days = 7
cnt = days * 96
title = '{}, {} Tage'
fig = plt.figure()
fig.set_size_inches(30,10)
fig.suptitle(title.format(data, days), fontsize=20)
plt.plot(np.arange(0,len(df_data[data][0:cnt])), df_data[data][0:cnt])
```
